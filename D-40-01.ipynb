```python
#splitting the merged attributers into the two differnt attributes
stats1['job']=stats1.jobedu.apply(lambda x: x.split(',')[0] )
stats1['edu']=stats1.jobedu.apply(lambda x: x.split(',')[1] )
stats1
```

```python
stats1.drop(['jobedu'], axis=1,inplace=True)
stats1.drop(['poutcome'],axis=1,inplace=True)
```

```python
stats1.isnull().sum()
```

```python
stats1.shape
```

```python
#remove the rows where values are misssing
x=stats1.age.isnull().sum()
x
```

```python
stats1.dropna(subset=['age'],inplace=True)
stats1
```

```python
stats1.month.mode()
```

```python

stats1['month']=stats1['month'].fillna(stats1['month'].mode())
stats1
```

```python
#replacing specific entities in a row
stats1.loc[stats1["marital"]=="married","marital"]=0
stats1.loc[stats1["marital"]=="single","marital"]=1
stats1.loc[stats1["marital"]=="divorced","marital"]=2
stats1
```

```python
stats1.duration=stats1.duration.apply(lambda x: float(x.split()[0])/60 if x.find("sec")>0 else float(x.split()[0] ))

```
