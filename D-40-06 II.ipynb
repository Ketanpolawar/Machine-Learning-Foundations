```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,target,test_size=0.2)
```

```python
from sklearn import svm#support vector margin(hyperplane)
model=svm.SVC()
model.fit(x_train,y_train)
model.score(x_test,y_test)
y_prediction=model.predict(x_test)

```

```python

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_prediction,labels=model.classes_)
```

```python
from sklearn.metrics import classification_report#liner poly and rbf kernel (rbf does the infine dimension ) ,gamma will decide grouping, c define decision boundries
print(classification_report(y_test,y_prediction))
```
